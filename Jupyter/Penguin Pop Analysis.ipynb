```python
# Creating datasets of male and female penguins
male = penguins[penguins['sex'] == 'Male']
female = penguins[penguins['sex'] == 'Female']
```

```python
male
```

```python
female
```

```python
# Grouping on island, followed by species, males
male_grouping = male.groupby(['island', 'species']).mean()
male_grouping.head()
```

```python
# Grouping on island, followed by species, females
female_grouping = female.groupby(['island', 'species']).mean()
female_grouping.head()
```

```python
f_island_pops = female.groupby('island')['sex'].count()
# 80 females on Biscoe
# 61 females on Dream
# 24 females on Torgensen
f_island_pops
```

```python
m_island_pops = male.groupby('island')['sex'].count()
# 83 males on Biscoe
# 62 males on Dream
# 23 males on Torgensen
m_island_pops
```

```python
total_pops = pd.merge(f_island_pops, m_island_pops, how='inner', right_index=True, left_index=True)
total_pops
```

```python
total_pops['island_pop'] = total_pops['sex_x'] + total_pops['sex_y']
total_pops
```

```python
total_pops["female_pop_percentage"] = (total_pops["sex_x"]/total_pops["island_pop"])*100
total_pops["males_pop_percentage"] = (total_pops["sex_y"]/total_pops["island_pop"])*100
total_pops
```

```python
frames = [male_grouping, female_grouping]
frames
```

```python
result = pd.concat(frames, keys = ['Male', 'Female'])
result
```

```python
total_pops['island_percent'] = (total_pops['island_pop']/total_pops.island_pop.sum())*100
total_pops
```

```python
biscoe_percent = round(total_pops.island_percent['Biscoe'], 2)
biscoe_percent
```

```python
dream_percent = round(total_pops.island_percent['Dream'], 2)
dream_percent
```

```python
torgersen_percent = round(total_pops.island_percent['Torgersen'], 2)
torgersen_percent
```

```python
labels = 'Biscoe', 'Dream', 'Torgersen'
```

```python
sizes = [biscoe_percent, dream_percent, torgersen_percent]
```

```python
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Penguin Population per Island')
```

## Tips for comparing islands
Don't group together Islands AND sex, instead group only by Island

That way, you can add 'Sex' as a hue, and start getting multiple bars and stacks

```python
penguins = penguins.dropna(axis=0)
penguins
```

```python
island_group = penguins.groupby('island')
island_group.mean().head()
```

```python
island_group.head()
```

```python
# Based on the above data alone, we can plot:
# Heaviest penguins per island (and also per species?)
# Average bill or flipper properties per island (also per species?)
```

```python
# need to group by island or species, try adding a column chart for mass
```

```python
# Mean penguin mass per island
sns.catplot(x=island_group.mean().index, y='body_mass_g', data=island_group.mean(), kind='bar')
plt.title('Mean Penguin Mass per Island')
plt.xlabel('Island')
plt.ylabel('Body Mass (g)')
```
